```python
import cv2
```

```python
up_cascade = cv2.CascadeClassifier('data/up/cascade.xml')
right_cascade = cv2.CascadeClassifier('data/right/cascade.xml')
down_cascade = cv2.CascadeClassifier('data/down/cascade.xml')
left_cascade = cv2.CascadeClassifier('data/left/cascade.xml')
```

```python
font = cv2.FONT_HERSHEY_COMPLEX
#up - 1
#right - 2
#down - 3
#left - 4
```

```python
def detect_arrow(cascade, gray, img):
    obj = cascade.detectMultiScale(gray, scaleFactor=1.6, minNeighbors=9, minSize=(150, 150))
    v = 0
    for (x, y, w, h) in obj:
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,255,0),2)
        v = y+h
    return v
```

```python
cap = cv2.VideoCapture(0)
prev = 1
curr = 1
while True:
    ret, img = cap.read()
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    up = detect_arrow(up_cascade, gray, img)
    right = detect_arrow(right_cascade, gray, img)
    down = detect_arrow(down_cascade, gray, img)
    left = detect_arrow(left_cascade, gray, img)
    
    if(up > 0):
        cv2.putText(img, 'UP', (0, 130), font, 1, (0, 255, 255), 2, cv2.LINE_AA)
        curr = 1
    elif(right > 0):
        cv2.putText(img, 'RIGHT', (0, 130), font, 1, (0, 255, 255), 2, cv2.LINE_AA)
        curr = 2
    elif(down > 0):
        cv2.putText(img, 'DOWN', (0, 130), font, 1, (0, 255, 255), 2, cv2.LINE_AA)
        curr = 3
    elif(left > 0):
        cv2.putText(img, 'LEFT', (0, 130), font, 1, (0, 255, 255), 2, cv2.LINE_AA)
        curr = 4
    
#     text = "CONSTANT"
    if(curr - prev != 0):
        if(curr - prev == 1):
            text = "CLOCKWISE"
        elif(curr - prev == -1):
            text = "ANTI-CLOCKWISE"
    
    cv2.putText(img, text, (130, 130), font, 1, (0, 0, 255), 2, cv2.LINE_AA)
            
    prev =  curr
    cv2.imshow('img',img)
    k = cv2.waitKey(30) & 0xff
    if k == 27:
        break
        
cap.release()
cv2.destroyAllWindows()        
```
